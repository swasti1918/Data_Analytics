```python
import numpy as np
import pandas as pd
```

```python
# Finding missing data

data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
df
```

```python
df.isna()   ## for finding null values(nan)
```

```python
df.isna().sum()  ### total count of null values
```

```python
df.isna().any()  # sbme null values h ya koi khaali bhi h
```

```python
## Removing missing data (based on rows)

df.dropna()   ## for removing null values of rows
```

```python
df.dropna(thresh=3)  ## 3 means u have at least 3 non null values
```

```python
## Filling missing data

df.fillna(0)
```

```python
values = {'A':0,'B':100,'C':300,'D':400}  #based on col wise
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
