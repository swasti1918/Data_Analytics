```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

***Matplotlib Basics***

```python
x = np.linspace(0,5,11)  #this means 0 and 5 ke beech me 11 elements chahiye
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

***Subplots in Matlplotlib***

```python
plt.subplot(2,2,1)   # this means a square and there is 4 boxes in that and the 3rd no shows the box num
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)  # i can plot anything whatever i want
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(y,x) 
```

***Object-Oriented way***

```python
fig = plt.figure(figsize=(20,8),dpi=100)
axis1 = fig.add_axes([0.1,0.1,1,1])  # we can provide values in b/w 0 - 1
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.1,0.3,0.3])
axis2.plot(x,y)
```

***Types of Plots***

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

***Saving Plots***

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
