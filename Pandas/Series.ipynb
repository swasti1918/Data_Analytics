```python
# Creating Series (there are multiple way to create pandas series)

import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c','d']
my_list = [1,2,3,4]
arr = np.array([1,2,3,4])
dict = {'a':10,'b':20,'c':30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)  # this means we are giving custom indexes
```

```python
pd.Series(arr)
```

```python
pd.Series(dict)
```
