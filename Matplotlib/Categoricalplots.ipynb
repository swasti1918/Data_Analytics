```python
import matplotlib.pyplot as plt
import seaborn as sns
```

***Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

boxplot
violinplot
stripplot
swarmplot
barplot
countplot
Let's go through examples of each!***

```python
df = sns.load_dataset("tips")
df
```

```python
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
sns.barplot(x=df['sex'], y = df['total_bill'])
```

```python
import numpy as np
sns.barplot(x=df['sex'], y = df['tip'],estimator = np.sum)
```

```python
sns.boxplot(x = 'tip', y = 'day', data = df, palette = 'rainbow')  #boxplot means comparing categorial data with quantitative data
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.stripplot(x = 'day', y = 'total_bill',data = df) #scatter form stripplot
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df)
```
