```python
import numpy as np
import pandas as pd
```

***Group By***

```python
# DataFrame
data = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S1', 'S2'],
    'Sales': [100, 200, 150, 250, 120, 180, 130, 220],
    'Quantity': [10, 15, 12, 20, 8, 18, 11, 25],
    'Date': pd.date_range(start='2023-01-01', periods=8)
})
df = pd.DataFrame(data)
df
```

```python
# Group by category and calculate sum of sales

cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
# Group by store and calculate sum of sales

cat = df.groupby('Store')['Sales'].sum()
cat
```

```python
# Group by multiple colums
# Group by category and store

cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```
