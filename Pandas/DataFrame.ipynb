```python
import numpy as np
import pandas as pd
```

```python
## Creating DataFrame


data = {
    'Name': ['Aman', 'Riya', 'Karan', 'Sneha'],
    'Age': [31, 21, 33, 22],
    'City': ['Delhi', 'Mumbai', 'Paris', 'Bangalore'],
    'Salary': [30000, 35000, 40000, 38000]
}

df = pd.DataFrame(data)

```
